```python
import pandas as pd
import sqlalchemy

Server = "localhost"
Database = "Turbo_AZ"
Driver = "ODBC Driver 17 for SQL Server" # in order to get this info, click Win + S and write 'odbc', then open drivers menu and scroll down
engine = sqlalchemy.create_engine(f'mssql://@{Server}/{Database}?driver={Driver}')
```

```python
tbl = pd.read_sql('SELECT * FROM Cars', engine)
tbl
```

```python
df = tbl.drop(['ID'], axis = 'columns')
df
```

```python
USD_AZN = converter(1, 'USD', 'AZN')
EUR_AZN = converter(1, 'EUR', 'AZN')
def convert_to_AZN(x): 
    x = x.replace(' ', '')
    amount = x[:-3]
    if x.endswith('USD'):
        return float(amount) * USD_AZN
    elif x.endswith('EUR'):
        return  float(amount) * EUR_AZN
    return float(amount)
```

```python
df['Price_azn'] = df['Price'].apply(convert_to_AZN)
df[df['Model'] == 'Maserati Grecale']
```
