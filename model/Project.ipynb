```python
import pandas as pd
import numpy as np
from CurrencyConverter import converter

import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
import sqlalchemy

Server = "localhost"
Database = "Turbo_AZ"
Driver = "ODBC Driver 17 for SQL Server" # in order to get this info, click Win + S and write 'odbc', then open drivers menu and scroll down
engine = sqlalchemy.create_engine(f'mssql://@{Server}/{Database}?driver={Driver}')
```

```python
tbl = pd.read_sql('SELECT * FROM Cars', engine)
tbl
```

```python
#tbl.to_csv('cars.csv', index = False) 
```

```python
df = tbl.drop(['ID'], axis = 'columns')
df
```

```python
USD_AZN = converter(1, 'USD', 'AZN')
EUR_AZN = converter(1, 'EUR', 'AZN')
def convert_to_AZN(x): 
    x = x.replace(' ', '')
    amount = x[:-3]
    if x.endswith('USD'):
        return float(amount) * USD_AZN
    elif x.endswith('EUR'):
        return  float(amount) * EUR_AZN
    return float(amount)
```

```python
df1 = df.dropna()
df1
```

```python
df1['Price_azn'] = df1['Price'].apply(convert_to_AZN)
df1
```

```python
df1.drop(['Price'], axis = 'columns', inplace = True)
```

```python
df1.describe()
```

```python
name_stats = df1.groupby('Model')['Model'].agg('count').sort_values(ascending = False)
name_stats
```

```python
name_stats_less_than_5 = name_stats[name_stats < 5]
name_stats_less_than_5
```

```python
df1['Model'] = df1['Model'].apply(lambda x: 'other' if x in name_stats_less_than_5 else x)
df1[df1['Model'] == 'other']
```

```python
df1
```

```python
df2 = pd.get_dummies(data = df1, columns = ['Model'])
df2
```

```python
df3 = df2.drop(['Model_other'], axis = 'columns')
df3
```

```python
cols_to_dummy = ['Engine_Type', 'Gearbox', 'Transmitter']
df4 = pd.get_dummies(data = df3, columns = cols_to_dummy, drop_first = True)
df4
```

```python
df4.replace({True : 1, False : 0}, inplace = True)
df4
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
cols_to_scale = ['Year', 'Engine_Volume', 'Horse_Power', 'Mileage']
df4[cols_to_scale] = scaler.fit_transform(df4[cols_to_scale])
df4
```

```python
x = df4.drop(['Price_azn'], axis = 'columns')
y =  df4['Price_azn']
```

```python
from sklearn.preprocessing import Polynomial
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
clf_linear_reg = LinearRegression()
clf_linear_reg.fit(x_train, y_train)
clf_linear_reg.score(x_test, y_test)
```

```python
from sklearn.linear_model import Lasso
clf_lasso = Lasso()
clf_lasso.fit(x_train, y_train)
clf_lasso.score(x_test, y_test)
```

```python
from sklearn.linear_model import Ridge
clf_ridge = Ridge()
clf_ridge.fit(x_train, y_train)
clf_ridge.score(x_test, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
clf_rnd_forest = RandomForestRegressor(n_estimators = 200, random_state = 42)
clf_rnd_forest.fit(x_train, y_train)
clf_rnd_forest.score(x_test, y_test)
```
