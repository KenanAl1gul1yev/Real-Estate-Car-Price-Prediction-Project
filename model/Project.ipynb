```python
tbl = pd.read_sql('SELECT * FROM Cars', engine)
tbl
```

```python
df = tbl.drop(['ID'], axis = 'columns')
df
```

```python
USD_AZN = converter(1, 'USD', 'AZN')
EUR_AZN = converter(1, 'EUR', 'AZN')
def convert_to_AZN(x): 
    x = x.replace(' ', '')
    amount = x[:-3]
    if x.endswith('USD'):
        return float(amount) * USD_AZN
    elif x.endswith('EUR'):
        return  float(amount) * EUR_AZN
    return float(amount)
```
